```python
import pandas as pd
df=pd.read_csv("sample_2.csv", parse_dates=["OrderDate"])
df
```

```python
new_df=df.fillna(4.99)
new_df
```

```python
new_df=df.fillna({
    "Rep" : "Jones",
    "UnitCost" : "4.99" ,
    "Total" : "sum"
})
new_df
```

```python
new_df=df.fillna(method="ffill")
new_df
```

```python
new_df=df.fillna(method="bfill",axis="columns")
new_df
```

```python
new_df=df.fillna(method="ffill", limit=1)
new_df
```

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.dropna()
new_df
```

```python
import pandas as pd
df=pd.read_csv("sample_2.csv",nrows=10, parse_dates=["OrderDate"])
df
```

```python
new_df=df.dropna()
new_df
```

```python
import numpy as np
new_df=df.replace(1.99,np.NaN)
new_df
```

```python
new_df=df.replace({
    1.99:np.NaN,
    'Pencil': 'eraser'
})
new_df
```

```python
new_df=df.replace({
    'Region':'[A-Za-z]'
},'',regex=True)
new_df
```
